2. Count blurry images

```python
import os
import json

json_file_path = "../BlurDetection2/train_blurr_result.json"


with open(json_file_path, 'r') as file:
    json_data = json.load(file)

json_data['results']

blurry_images = []
normal_images =[]
for data in json_data['results']:
    if data['blurry'] == True:
        blurry_images.append(os.path.basename(data['input_path']))
        
    else:
        normal_images.append(os.path.basename(data['input_path']))

total = len(blurry_images) + len(normal_images)
percent_blurry = round(len(blurry_images)*100 / total, 2)
percent_normal = round(len(normal_images)*100 / total, 2)



print("precent of images", '\n', 'percent_blurry:', percent_blurry, 'percent_normal', percent_normal)
```

3. Sort blurry images to folders

```python
import os
import json
import shutil
from tqdm import tqdm

json_file_path = "../BlurDetection2/train_blurr_result.json"
normal_folder = '../../dataset/train_normal'
blurry_folder = '../../dataset/train_blurry'
source_directory = "../../dataset/train"

os.makedirs(normal_folder, exist_ok=True)
os.makedirs(blurry_folder, exist_ok=True)

with open(json_file_path, 'r') as file:
    json_data = json.load(file)

json_data['results']

# Copy Blurry 
for image in tqdm(blurry_images):
    source_path = os.path.join(source_directory, image)
    target_path = os.path.join(blurry_folder, image)
    shutil.copy2(source_path, target_path)

# Copy Normal
for image in tqdm(normal_images):
    source_path = os.path.join(source_directory, image)
    target_path = os.path.join(normal_folder, image)
    shutil.copy2(source_path, target_path)
```
